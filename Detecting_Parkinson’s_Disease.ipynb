```python
#train-test data split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =7)
```

```python
X_train.shape, X_test.shape
```

```python
#importing different metrics
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, classification_report, roc_curve, auc
```

```python
#function to calculate train and test accuracy
def accuracy(model, X_train, X_test, y_train, y_test):
  y_train_pred  = model.predict(X_train)
  train_accuracy = accuracy_score(y_train, y_train_pred)

  y_test_pred = model.predict(X_test)
  test_accuracy = accuracy_score(y_test, y_test_pred)

  return train_accuracy, test_accuracy
```

##Trying out different models

#SVM

```python
#Train the model
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
#Train & Test accuracy of SVM
accuracy(model, X_train, X_test, y_train, y_test)
```

#Logistic Regression

```python
#Train the model
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(solver='liblinear')
model1.fit(X_train,y_train)
```

```python
#Train & Test accuracy of LR
accuracy(model1, X_train, X_test, y_train, y_test)
```

#KNN

```python
#Train the model
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors = 3)
model2.fit(X_train,y_train)
```

```python
#Train & Test accuracy of KNN
accuracy(model2, X_train, X_test, y_train, y_test)
```

#Decision Tree

```python
#Train the model
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier(max_depth = 6,min_samples_split = 2,max_leaf_nodes = 2)
model3.fit(X_train,y_train)
```

```python
#Train & Test accuracy of Decision Tree
accuracy(model3,X_train, X_test, y_train, y_test)
```

#Random Forest

```python
#Train the model
from sklearn.ensemble import RandomForestClassifier

model4 = RandomForestClassifier(max_depth = 3,n_estimators = 1000,criterion = 'gini')
model4.fit(X_train,y_train)
```

```python
#Train & Test accuracy of Decision tree
accuracy(model4,X_train, X_test, y_train, y_test)
```

```python
#Precision, recall & F1 scores obtained with random forest
print(classification_report(y_test, model4.predict(X_test)))
```
